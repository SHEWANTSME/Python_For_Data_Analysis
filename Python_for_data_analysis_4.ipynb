#### List Comprehension

````python
print([i for i in range(10)])# 리스트 컴프리헨션
print([i+2 for i in range(10)])
print([i*i for i in range(10)])
print([i**2 for i in range(10)])
````

````python
print([i-j for i in range(1,5) for j in range(1,6)])# 이중 for문
# 1-1,1-2,1-3,1-4,1-5,2-1,2-2,2-3,2-4,2-5,3-1,3-2,3-3,3-4,3-5,4-1,4-2,4-3,4-4,4-5
# 5*4=20개 -> j가 5번 돌아야 i가 바뀜

# 위의 list comprehension은 아래와 같다
a= []
for i in range(1,5):
    for j in range(1,6):
        a+=[i-j] # a.append(i-j)와 같다
print(a)
````

````python
# 이런식으로 이중 for문의 원소를 채울 수 있음
a= [[i for i in range(1,10)] for j in range(1,5)]
# 4 by 9 행렬
print(a)
````

````python
# 아예 같은 숫자로 채울꺼면
a=[[1 for _ in range(9)] for _ in range(4)]
print(a)
# 1로 채운다면, numpy에선 np.ones((4,9))로 쓸 수 있음
import numpy as np
print(np.ones((4,9))) # 4 by 9 행렬
````

````python
# deepcopy()와 copy()의 차이는
# copy()는 얕은 복사, deepcopy()는 깊은 복사
# 얕은 복사는 리스트 안의 리스트까지 복사하지 않음
# 예시를 들어보자
a = [[1,2,3],[4,5,6]]
b= a.copy()
print(b)
a[0].append(30)
print(b) # a를 바꾸면 b도 바뀐다. -> 얕은 복사
print(a)
print(a is b)# a랑 b는 다른 메모리를 가리키고 있음
print(a == b)# a랑 b는 같은 값을 가지고 있어서 True
````

````python
# 깊은 복사를 해보자
a = [[1,2,3],[4,5,6]]
# b = a.deepcopy() deepcopy()는 import copy-> copy 라이브러리를 써야함
import copy
b=copy.deepcopy(a)
a[0].append(30)
print(b) 
print(a)
print(a is b)
print(a == b)
````

````python
dic = {"a":10, "b":20, "c":30}
dic.setdefault("d")# setdefault는 키가 없으면 추가해줌, value를 지정하지 않으면 None으로 지정
dic.setdefault("e",50)# key를 e로 하고 value를 50으로 지정
dic["f"] = None # 이렇게 지정해줘도 됨
print(dic)
````
