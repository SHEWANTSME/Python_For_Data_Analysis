```python
# 누락된 data처리
import pandas as pd
import numpy as np
```

```python
string_data = pd.Series(['alpha' , 'beta' , np.nan , 'gamma'])
string_data
```

```python
data = pd.Series([1,np.nan , 3.5,np.nan, 7])
data.dropna() # np.nan(누락된 data)를 골라내줌  == data[data.notnull()]
```

```python
data[data.notnull()]
```

```python
# dataframe에서는 nan값을 하나라도 포함하고 있는 row를 제외시킴
data = pd.DataFrame([[1,3,5] , [1,np.nan, np.nan] , [np.nan,3,2] , [np.nan,np.nan , np.nan]])
data
```

```python
data1 = data.dropna()
data1
```

```python
# how= 'all' 이면 싹 다 nan 인거 빼곤 놔둠
data2 = data.dropna(how = 'all')
data2
```

```python
# column을 제외하는것도 동일함.
data[4] = np.nan
data
```

```python
data.dropna(axis = 1, how = 'all')
```

```python
# 아 그리고 행에서 nan이 아닌놈을 n개 이상 보고싶으면 thresh = n 을 조건 추가하면 된다
data.dropna(thresh=2)
```

```python
# 결측치 채우기
data.fillna(1000) # 이런식으로 넣고싶은 값을 채워주는게 fillna
```

```python
# dict값으로 column마다 다른값 채우는것도 가능함
data.fillna({0:'뀨잉' ,1 : '떡상이' })
```

```python
# fillna는 새로운 객체를 반환하지만 기존객체에 변경을 원하면 inplace = True를 집어넣어주면 된다
data.fillna({0:"뀨뀨뀨" , 1 : "떡떡상상이"} , inplace = True)
```

```python
# iloc와 loc 차이점
# iloc = integer location = DF의 행이나 열 순서를 나타내는 정수로 특정값 추출하는 방법
# loc = location = 
mid=pd.read_csv('midwest.csv')
mid.head(10)
```

```python
mid.iloc[0] # 전체 dataframe에서 0번째 행에 있는 값만 추출해라
```

```python
mid.loc[0] # 전체 dataframe 에서 인덱스 이름이 0 인 행만 추출해라
```

```python
#mid.iloc[0,2]
mid.iloc[0][2]  = mid.iloc[0,2] = mid.loc[0][2]  = mid.loc[0,'state'] but not same as mid.loc[0,2]
```

```python
mid.loc[0,'state']
```

```python
mid.iloc[:300:-2 , :3] # 이것저것 만져보면 알듯!
```

```python
aa = mid.iloc[21:40 , 1:3]
```

```python
aa
```

```python
aa.loc[40] = ['JASPER','IL'] # 행추가 by using loc -> 여러개 넣고 싶으면 append
```

```python
aa
```

```python
# 중복제거 using duplicated
aa.duplicated()
```

```python
aa.drop_duplicates() # 없어진걸 확인할 수 있다.
```
