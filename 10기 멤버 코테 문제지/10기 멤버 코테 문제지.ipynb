### 문제 [3-4] - 6점

* percent를 아래의 분류 기준에 따라 새로운 변수를 추가해주시고, 각 분류에 해당하는 지역이 몇 개인지 확인해보시오.

* large: 0.9% 이상 / middle: 0.4~0.9% 미만 / small: 0.4% 미만

```python

new_mid['percent_2'] = new_mid['category']



large_cnt = 0
mid_cnt = 0
small_cnt=0
total_cnt =-1
for i in new_mid['percent']:
    total_cnt+=1
    if(i>=0.9):
        large_cnt+=1
        new_mid['percent_2'][total_cnt] = 'large'
    elif(i>=0.4 and i<0.9):
        mid_cnt+=1
        new_mid['percent_2'][total_cnt] = 'middle'        
    else:
        small_cnt+=1
        new_mid['percent_2'][total_cnt] = 'small'
        
print(large_cnt , mid_cnt , small_cnt)
print(new_mid['percent_2'])
```
