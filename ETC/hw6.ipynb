```python
# (c)번 문제
# (c)-1 Column 0(임신 횟수)을 0회, 1 ~ 3회, 3회 이상의 3개 구간으로 나누고 
# (c)-2 one-hot encoding을 수행한 후, 
# (c)-3 중간 20행을 sparse matrix, numpy array, DataFrame 형태로 각각 출력하시오.
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
dataset = pd.read_csv("pima-indians-diabetes.csv", header = None)
df = pd.DataFrame(dataset)
for i in range(768): # (c)-1 부분 for문을 돌려서 해당되는 부분의 dataframe만 값을 변경했습니다.
    if 1<=df.loc[:767,0][i]<=3: # 0일땐 어차피 0 이니 신경안써도 되고, 1부터 3 사이값은 1로
        df.loc[:767,0][i] = 1
    elif df.loc[:767,0][i]>3: # 3보다 큰값일땐 2로 설정했습니다.
        df.loc[:767,0][i]= 2
        
# column name이 있는게 좋을것 같아서 써놨습니다. column name은 kaggle에서 가져왔습니다.
df.columns = ["Pregnancies","Glucose","BloodPressure","SkinThickness", "Insulin", "BMI" , "DiabetePedigreeFunc","Age", "Outcome"]

# one- hot encoding
onehot = pd.get_dummies(df["Pregnancies"] )
new_onehot = onehot.head(20) # 상위 20개를 new_onehot으로 설정했습니다. 
print(new_onehot) # 그냥 dataframe형태

sparse = OneHotEncoder().fit_transform(new_onehot)
print(sparse) # sparse matrix형태
ar = sparse.toarray()
print(ar) # numpy array형태 
```
