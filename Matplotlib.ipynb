```python
import numpy as np
# red, 별모양, dash선으로 sin 그리기
# green, 동그라미, 실선으로 cos 그리기
theta=np.linspace(0,2*np.pi,20)# 0부터 2파이까지 20개로 나누기
# np.linespace(start, end, num) : start부터 end까지 num개로 나누기
plt.plot(theta,np.sin(theta),"r*--",markersize=20, label="sin")# label은 범례(우상단 박스안의 내용)
# plt.figure() # 새창 열기. 이거 없으면 겹쳐 그리기
# plt.figure() : 새로운 figure 창을 열어서 그림을 그림
plt.plot(theta,np.cos(theta),"go-",markersize=4, label="coss")# label은 범례(우상단 박스안의 내용)
# plt.plot(a,b,c,d,e) -> a: x축 값, b: y축 값, c: 선의 색깔, d: 마커사이즈, e: 선의 모양
# 마커 사이즈는 별, 동그라미의 사이즈
plt.xlabel("theta")# x축 이름
plt.ylabel("value")# y축 이름
plt.title("sin and cos",color="b")# 그래프 이름
plt.legend()# legend는 우상단 박스자체
plt.grid()# 격자
# plt.xlim([0, np.pi])
# plt.ylim([-2, 2])
plt.axis([0, np.pi, -2,2])# x축, y축 범위
# 범위 설정 안하면 default로 설정됨
```

```python
plt.plot(theta,np.sin(theta),"r*--",markersize=10, label="sin")# 
plt.plot(theta,np.cos(theta),"go-",markersize=10, label="cos")
plt.xlabel("theta",fontsize=20)# 폰트사이즈 수정 가능
plt.ylabel("value",fontsize=20)
plt.title("sin and cos",color="b",fontsize=20)
plt.legend(fontsize=20)
plt.grid()
plt.xticks(fontsize=5)# x축 숫자들의 폰트사이즈
plt.yticks(fontsize=10)# y축 숫자들의 폰트사이즈
```

```python
x=range(5)# (0,1,2,3,4)
y1=[i for i in range(1,6)]# y1=[1,2,3,4,5]
y2=[i+100 for i in range(1,6)]# y2=[101,102,103,104,105]

plt.figure()# 그래프를 그리기 전에 흰 도화지를 깐다고 생각하셈
plt.plot(x,y1)# x축은 0,1,2,3,4, y축은 y1
plt.plot(x,y2)# x축은 0,1,2,3,4, y축은 y2

plt.figure()# 새로운 도화지
fig, ax1 = plt.subplots()# 
ax2 = ax1.twinx()# x축을 공유하는 새로운 y축을 만듦
p1=ax1.plot(x,y1,'b')# x축은 0,1,2,3,4, y축은 y1, 파란색
p2=ax2.plot(x,y2,'r')# x축은 0,1,2,3,4, y축은 y2, 빨간색
ax1.set_ylim([-10, 10])# y1의 y축 범위(redline)
ax1.set_xlabel("x")# x축 이름
ax1.set_ylabel("y1")# y1의 y축 이름
ax2.set_ylabel("y2")# y2의 y축 이름
ax1.set_title("y1 and y2")# 그래프 이름
ax1.grid()
plt.legend(p1+p2,["y11","y2"], loc="right")# label을 legend에서 설정할 수 있음
```

```python
a=np.linspace(-1,1,100)# -1부터 1까지 100개로 나누기
b=np.linspace(-1,1,100)# -1부터 1까지 100개로 나누기
A, B=np.meshgrid(a,b)# a,b의 모든 조합을 만들어줌
# meshgrid는 2차원 그리드 포인트를 만들어줌
# numpy의 meshgrid 함수는 1차원 좌표 배열 (x1, x2..... xn)에서 N차원 직사각형 격자를 만드는 함수
L = A**2 + B**2# L은 a^2+b^2 -> 원
plt.contour(a,b,L,20) # 20 -> 등고선 개수
# plt.contour() 하면 plt.figure()를 안해도 됨 -> 어차피 contour는 새로운 창을 열어서 그림
plt.xlabel('a'); plt.ylabel('b'); plt.grid()# x축 이름, y축 이름, 격자
# 근데 두번째거를 그릴려면 plt.figure()를 해야함
plt.figure()
plt.contourf(a,b,L,5)# contourf는 새로운 색깔로 채워진 등고선
```

```python
fig, ax = plt.subplots(figsize=[10, 9])
# plt.subplots는 figure와 axes를 반환
CS = ax.contour(a,b,L,20) # 20 -> 등고선 개수
ax.clabel(CS, fontsize=10)
plt.xlabel('a'); plt.ylabel('b'); plt.grid()
```

```python
import plotly.graph_objects as go

a=np.linspace(-2,2,100)
b=np.linspace(-2,2,100)
A, B = np.meshgrid(a,b)
L = np.cos(A**2.4)+np.sin(B**2.4)

fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, colorscale="viridis", opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True))
fig.update_layout(title='cos(A^2.4)+sin(B^2.4)', width=700, height=600)
```
