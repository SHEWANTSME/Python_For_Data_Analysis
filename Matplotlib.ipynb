```python
import numpy as np
# red, , dash sin 
# green, ,  cos 
theta=np.linspace(0,2*np.pi,20)# 0 2 20 
# np.linespace(start, end, num) : start end num 
plt.plot(theta,np.sin(theta),"r*--",markersize=20, label="sin")# label (  )
# plt.figure() #  .    
# plt.figure() :  figure    
plt.plot(theta,np.cos(theta),"go-",markersize=4, label="coss")# label (  )
# plt.plot(a,b,c,d,e) -> a: x , b: y , c:  , d: , e:  
#   ,  
plt.xlabel("theta")# x 
plt.ylabel("value")# y 
plt.title("sin and cos",color="b")#  
plt.legend()# legend  
plt.grid()# 
# plt.xlim([0, np.pi])
# plt.ylim([-2, 2])
plt.axis([0, np.pi, -2,2])# x, y 
#    default 
```

```python
plt.plot(theta,np.sin(theta),"r*--",markersize=10, label="sin")# 
plt.plot(theta,np.cos(theta),"go-",markersize=10, label="cos")
plt.xlabel("theta",fontsize=20)#   
plt.ylabel("value",fontsize=20)
plt.title("sin and cos",color="b",fontsize=20)
plt.legend(fontsize=20)
plt.grid()
plt.xticks(fontsize=5)# x  
plt.yticks(fontsize=10)# y  
```

```python
x=range(5)# (0,1,2,3,4)
y1=[i for i in range(1,6)]# y1=[1,2,3,4,5]
y2=[i+100 for i in range(1,6)]# y2=[101,102,103,104,105]

plt.figure()#       
plt.plot(x,y1)# x 0,1,2,3,4, y y1
plt.plot(x,y2)# x 0,1,2,3,4, y y2

plt.figure()#  
fig, ax1 = plt.subplots()# 
ax2 = ax1.twinx()# x   y 
p1=ax1.plot(x,y1,'b')# x 0,1,2,3,4, y y1, 
p2=ax2.plot(x,y2,'r')# x 0,1,2,3,4, y y2, 
ax1.set_ylim([-10, 10])# y1 y (redline)
ax1.set_xlabel("x")# x 
ax1.set_ylabel("y1")# y1 y 
ax2.set_ylabel("y2")# y2 y 
ax1.set_title("y1 and y2")#  
ax1.grid()
plt.legend(p1+p2,["y11","y2"], loc="right")# label legend   
```

```python
a=np.linspace(-1,1,100)# -1 1 100 
b=np.linspace(-1,1,100)# -1 1 100 
A, B=np.meshgrid(a,b)# a,b   
# meshgrid 2   
# numpy meshgrid  1   (x1, x2..... xn) N    
L = A**2 + B**2# L a^2+b^2 -> 
plt.contour(a,b,L,20) # 20 ->  
# plt.contour()  plt.figure()   ->  contour    
plt.xlabel('a'); plt.ylabel('b'); plt.grid()# x , y , 
#    plt.figure() 
plt.figure()
plt.contourf(a,b,L,5)# contourf    
```

```python
fig, ax = plt.subplots(figsize=[10, 9])
# plt.subplots figure axes 
CS = ax.contour(a,b,L,20) # 20 ->  
ax.clabel(CS, fontsize=10)
plt.xlabel('a'); plt.ylabel('b'); plt.grid()
```

```python
import plotly.graph_objects as go

a=np.linspace(-2,2,100)
b=np.linspace(-2,2,100)
A, B = np.meshgrid(a,b)
L = np.cos(A**2.4)+np.sin(B**2.4)

fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, colorscale="viridis", opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True))
fig.update_layout(title='cos(A^2.4)+sin(B^2.4)', width=700, height=600)
```
